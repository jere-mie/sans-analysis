```python
fname = '0aToc-6a-15c_merged.bin'
with open(fname, 'r') as f:
    contents = [i.strip().split('\t') for i in f.readlines()[2:]]
contents = [[float(j) for j in i] for i in contents]
x, y, e, dx = (list(x) for x in list(zip(*contents)))
```
